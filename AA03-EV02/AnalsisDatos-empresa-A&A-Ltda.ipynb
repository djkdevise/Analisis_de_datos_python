# Agrupamiento de datos

```python
- Obtener 6 rangos en una preagrupacion para poder agrupar los datos y analizarlos
```

```python
#Preagrupacion

# Calcula el rango total de precios
precio_min = df['Precio'].min()
precio_max = df['Precio'].max()

# Divide el rango total en 6 intervalos
intervalos = pd.cut(df['Precio'], bins=6)

# Crea una nueva columna con el rango de precios
df['Rango_Precio'] = intervalos

# Agrupa y cuenta la cantidad de registros en cada rango
conteo_rangos = df.groupby('Rango_Precio').size()

# Muestra los resultados
print(conteo_rangos)
```

```python
#Creacion de Variables

#1) Definir Rango de Edades
rangos=[1595087.12,2307503500.0,6891814460.0,9183969940.0,11476125420.0,13768280900.0]

#2) Establecer el nombre para cada rango
nombreRango=["A","B","C","D","E"]

#3) Se crea nueva columna con el nombre Rango_Edad donde almacenara le dato del rango de edad segun correponda 

df['Rango_Precio'] = pd.cut(df['Precio'], rangos, labels=nombreRango)

# Comprobamos que se ha creado una nueva columna
df.info()
```

```python
import pandas as pd

# Distribución porcentual de Ciudad
propiedades_por_ciudad = df['Ciudad'].value_counts()

# Crear un DataFrame con las etiquetas y los valores
data = {'Ciudad': propiedades_por_ciudad.index, 'Cantidad': propiedades_por_ciudad.values}
df_propiedades_por_ciudad = pd.DataFrame(data)

# Calcular el porcentaje de cada ciudad
total_propiedades = df_propiedades_por_ciudad['Cantidad'].sum()
df_propiedades_por_ciudad['Porcentaje'] = (df_propiedades_por_ciudad['Cantidad'] / total_propiedades) * 100

# Mostrar la tabla
print(df_propiedades_por_ciudad)
```
